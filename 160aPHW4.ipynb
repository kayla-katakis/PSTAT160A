```python
import numpy as np
import math
import matplotlib.pyplot as plt

from scipy.stats import arcsine

# Allows to render plots directly within the notebook
%matplotlib inline
```

```python
#Problem 1
def randomWalk(lengthOfRandomWalk):
    S_0 = 0
    samplePath = np.random.sample(lengthOfRandomWalk-1)
    for i in range(len(samplePath)):
        if samplePath[i] > 0.5:
            samplePath[i] = 1
        else:
            samplePath[i] = -1
    samplePath = np.cumsum(np.insert(samplePath,0,S_0), dtype = int)
    return samplePath
                   
```

```python
#Problem 2
def timeAboveZero(path):
    result = []
    for i in range(len(path)):
        if path[i] >0:
            result.append(path[i])
    result = len(result)
    return result

def TimeOfLastVisitOfZero(path):
    result = 0
    for i in range(len(path)):
        if path[i] == 0:
            result = i
    return result

def lastTimeHitMax(randomWalk):
    result = 0
    maximum = randomWalk[0]
    for i in range(len(randomWalk)):
        if randomWalk[i] >= maximum:
            maximum = randomWalk[i]
            result = i
    return result
```

```python
x = np.linspace(arcsine.ppf(0.05), arcsine.ppf(0.95), 100)
plt.title("Density of the arcsine distribution")
plt.plot(x, arcsine.pdf(x), linewidth=2, color='b')
```

```python
#Problem 3: Simulations
C2Nsim = []
L2Nsim = []
M2Nsim = []

for i in range(10000):
    C2Nsim.append(timeAboveZero(randomWalk(1000))/1000)
    L2Nsim.append(TimeOfLastVisitOfZero(randomWalk(1000))/1000)
    M2Nsim.append(lastTimeHitMax(randomWalk(1000))/1000)
                  
```

```python
#Problem 3: Histograms

c = C2Nsim  # This has to be replaced by the simulated values for C_2N

plt.figure(figsize=(10,5))
plt.title("Normalized histogram for 10000 realisations of $C_{1000}$")
plt.hist(c, bins='auto',density='true')
plt.plot(x, arcsine.pdf(x), linewidth=2, color='r', label="true arcsine density")
plt.legend()

l = L2Nsim  # This has to be replaced by the corresponding simulated values for L_2N

plt.figure(figsize=(10,5))
plt.title("Normalized histogram for 10000 realisations of $L_{1000}$")
plt.hist(l, bins='auto',density='true')
plt.plot(x, arcsine.pdf(x), linewidth=2, color='r', label="true arcsine density")
plt.legend()

m = M2Nsim  # This has to be replaced by the corresponding simulated values for M_2N

plt.figure(figsize=(10,5))
plt.title("Normalized histogram for 10000 realisations of $M_{1000}$")
plt.hist(m, bins='auto',density='true')
plt.plot(x, arcsine.pdf(x), linewidth=2, color='r', label="true arcsine density")
plt.legend()
```
